```python
X_train = pd.read_csv('../data/X_train.csv')
X_test = pd.read_csv('../data/X_test.csv')
y_train = pd.read_csv('../data/y_train.csv')
y_test = pd.read_csv('../data/y_test.csv')

print("✅ Data Siap! Mari kita latih XGBoost.")
```

```python
model_xgb = xgb.XGBRegressor(
    n_estimators=1000,
    learning_rate=0.05,
    max_depth=5,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    n_jobs=-1
)

print("Sedang melatih model...")
model_xgb.fit(
    X_train, y_train,
    eval_set=[(X_test, y_test)],
    verbose=False
)
print("✅ Selesai!")
```

```python
y_pred_xgb = model_xgb.predict(X_test)

mae = mean_absolute_error(y_test, y_pred_xgb)
rmse = np.sqrt(mean_squared_error(y_test, y_pred_xgb))
r2 = r2_score(y_test, y_pred_xgb)

print("\n" + "="*30)
print("HASIL AKHIR (XGBOOST):")
print("="*30)
print(f"MAE      : {mae:.4f} Ton/Hektar")
print(f"RMSE     : {rmse:.4f} Ton/Hektar")
print(f"R2 Score : {r2:.4f}")
print("="*30)
```

```python
print("\nMembuat Analisis SHAP...")
explainer = shap.Explainer(model_xgb)
shap_values = explainer(X_test)

plt.figure(figsize=(10, 6))
shap.summary_plot(shap_values, X_test, show=False)
plt.title("Fitur Terpenting Penentu Hasil Panen", fontsize=14)
plt.tight_layout()
plt.show()
```
