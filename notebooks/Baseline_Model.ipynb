```python
X_train = pd.read_csv('../data/X_train.csv')
X_test = pd.read_csv('../data/X_test.csv')
y_train = pd.read_csv('../data/y_train.csv')
y_test = pd.read_csv('../data/y_test.csv')

print("✅ Data berhasil dimuat!")
print(f"Jumlah Data Latih: {len(X_train)}")
print(f"Jumlah Data Uji: {len(X_test)}")
```

```python
print("\nSedang melatih model Decision Tree...")

dt_model = DecisionTreeRegressor(random_state=42)
dt_model.fit(X_train, y_train.values.ravel())

print("✅ Model selesai dilatih!")
```

```python
y_pred = dt_model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("-" * 30)
print("HASIL EVALUASI (BASELINE):")
print("-" * 30)
print(f"MAE (Rata-rata Meleset) : {mae:.4f} Ton/Hektar")
print(f"RMSE (Error Kuadrat)    : {rmse:.4f} Ton/Hektar")
print(f"R2 Score (Akurasi Pola) : {r2:.4f} (Maks 1.0)")
print("-" * 30)
```
